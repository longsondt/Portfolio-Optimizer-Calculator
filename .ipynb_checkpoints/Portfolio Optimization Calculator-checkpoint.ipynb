```python
import pandas as pd
import numpy as np
import pandas_datareader.data as web
import datetime as datetime
```

# Step 1. Setting up the necessary information 

```python
avaliable_fund = int(input('How much will you invest in this portfolio? '))
number_of_securities = int(input('How many securities will you have in this portfolio? '))

securities = []
for number in range(number_of_securities):
    securities.append(input(f'Please type in security {number+1}: ').upper())
securities
```

```python
print('Please enter a start date')
```

```python
start_year = int(input('Please enter a start year: '))
start_month = int(input('Please enter a start month: '))
start_day = int(input('Please enter a start date: '))
start_date = datetime.date(start_year, start_month, start_day)
start_date
```

```python
print('Please enter a end date')
```

```python
end_year = int(input('Please enter a end year: '))
end_month = int(input('Please enter a end month: '))
end_day = int(input('Please enter a end date: '))
end_date = datetime.date(end_year, end_month, end_day)
end_date
```

```python
df = pd.DataFrame()

for security in securities:
    df[security] = web.DataReader(security, data_source='yahoo', start = start_date, end = end_date)['Adj Close']

df
```

```python
pct_change = df.pct_change()
list(pct_change['FB'])
```
